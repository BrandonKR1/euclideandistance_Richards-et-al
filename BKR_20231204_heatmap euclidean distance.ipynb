```python
# Read excel files that contain an array of Pearson correlations calculated elsewhere (in our case, GraphPad Prism)

df_hM3Dq = pd.read_excel('hM3Dq proximity matrix.xlsx', index_col=None)
df_hM3Dq = df_hM3Dq.set_index('x')

df_mCherry = pd.read_excel('mCherry proximity matrix.xlsx', index_col=None)
df_mCherry = df_mCherry.set_index('x')
```

```python
# Plot hM3Dq matrix

color = sns.color_palette("rocket", as_cmap=True)
h = sns.clustermap(df_hM3Dq, cmap = color, vmin=0, vmax=8, method='complete',
               metric='euclidean',  linewidths=0, xticklabels=1, yticklabels=1)

# Edit dendrogram characteristics

from scipy.cluster import hierarchy
from scipy.cluster.hierarchy import dendrogram

hierarchy.dendrogram(h.dendrogram_col.linkage, color_threshold=13.0515, distance_sort=True,ax = (h.ax_col_dendrogram)) 

# Optional to add grid lines and y-axis scale to determine length of tree (comment out if needed). Change colour threshold number to change where the dendrogram is cut 

import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter, MaxNLocator
h.ax_col_dendrogram.axis('on')
h.ax_col_dendrogram.yaxis.set_major_locator(MaxNLocator())
h.ax_col_dendrogram.yaxis.set_major_formatter(ScalarFormatter())
h.ax_col_dendrogram.grid(axis='y', ls='--', color='grey')

plt.savefig('hM3Dq Euclidean.pdf')
```

```python
h = sns.clustermap(df_mCherry, cmap = color, vmin=0, vmax=8, method='complete',
               metric='euclidean', linewidths=0, xticklabels=1, yticklabels=1)

# Edit dendrogram characteristics

from scipy.cluster import hierarchy
from scipy.cluster.hierarchy import dendrogram

hierarchy.dendrogram(h.dendrogram_col.linkage, color_threshold=6, distance_sort=True,ax = (h.ax_col_dendrogram)) 

#Optional to add grid lines and y-axis scale to determine length of tree (comment out if needed). Change colour threshold number to change where the dendrogram is cut 

import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter, MaxNLocator
h.ax_col_dendrogram.axis('on')
h.ax_col_dendrogram.yaxis.set_major_locator(MaxNLocator())
h.ax_col_dendrogram.yaxis.set_major_formatter(ScalarFormatter())
h.ax_col_dendrogram.grid(axis='y', ls='--', color='grey')

plt.savefig('mCherry Euclidean.pdf')
```
